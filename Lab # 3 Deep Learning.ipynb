```python
import tensorflow as tf
import numpy 
```

```python
# Define variables
w = tf.Variable([0.2], dtype=tf.float32)  # initial weight
b = tf.Variable([-0.2], dtype=tf.float32)  # initial bias
```

```python
x = tf.constant([4, 11, 13, 34])  # input
y = tf.constant([2, 4, 6, 8])  # actual output
```

```python
# Define linear model
#linear_model = w * x + b
def linear_model(x, w, b):
    return w * x + b
```

```python

# Print initial values of weight and bias
print("Initial weight:", w.numpy())
print("Initial bias:", b.numpy())
```

```python
# Print predicted output
print("Predicted output is:")
predicted_output = linear_model(x,w,b)
print(predicted_output.numpy())
```

```python
squared_deltas = tf.square(linear_model(x,w,b) - y)  # used to square the error
loss = tf.reduce_sum(squared_deltas)  # reduce_sum is used to sum all the squared errors and give us rmse
print("loss or Error is ")
print(loss.numpy())  # evaluating the loss
```

```python
# Plot the data points
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [0, -1, -2, -3], label='Actual Data')
plt.plot([1, 2, 3, 4], predicted_output, color='red', label='Predicted Output')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Data Points and Predicted Output')
plt.legend()
plt.show()
```
